```python
import glob,os
import pandas as pd
import matplotlib.pyplot as plt
import torch
import seaborn as sns
```

```python
# args
METRIC_PATH = "../../experiments/mimic3/sept29/version_0/metrics.csv"
```

```python
def plot_val_metrics(path):
    df = pd.read_csv(path)
    sns.scatterplot(x="step",y="val_loss",data=df)
    plt.show()
    sns.scatterplot(x="epoch",y="val_loss",data=df)
    plt.show()
    sns.scatterplot(x="step",y="lr-Adam",data=df)
    plt.show()
    sns.scatterplot(x="step",y="epoch",data=df)
    plt.show()
```

```python
plot_val_metrics(METRIC_PATH)
```
