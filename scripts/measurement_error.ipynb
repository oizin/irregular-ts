```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/analysis.csv')
```

```python
df.loc[:,'dt'] = df.loc[:,'timer_dt'] - df.loc[:,'timer'] 
```

```python
plt.hist(df.loc[:,'dt'])
```

```python
df.loc[:,'change_glc']  = df.loc[:,'glc'] - df.loc[:,'glc_dt']
```

```python
plt.hist(df.loc[df.dt > 1,'change_glc'],bins=40)
plt.show()
```

```python
df.loc[df.dt < 0.01,:].change_glc.std()
```

```python
df.loc[df.dt < 0.01,:].change_glc.mad()
```

```python
df.loc[(df.dt < 0.01) & df.dt > 0,:].plot(x='change_glc',y='glc',kind='scatter')
```

```python
df.loc[(df.change_glc == 0) & df.dt > 0,'dt'].plot(x='dt',kind='hist',bins=30)
```
